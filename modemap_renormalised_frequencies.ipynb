```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
temperature_mode1 = []
frequency_mode1 = [] 
with open('/Users/w21013885/mode1_X_point_real_supercell_112/t-vs-omega.dat', 'r') as mode1:
    for line in mode1:
        line=line.split()
        if line[0] != '#': 
            temperature_mode1.append(float(line[0]))
            frequency_mode1.append(float(line[1]))

temperature_mode2 = []
frequency_mode2 = [] 
with open('/Users/w21013885/mode2_X_point_real_supercell_112/t-vs-omega.dat', 'r') as mode1:
    for line in mode1:
        line=line.split()
        if line[0] != '#': 
            temperature_mode2.append(float(line[0]))
            frequency_mode2.append(float(line[1]))
```

```python
plt.plot(temperature_mode1,frequency_mode1, label='mode1')
plt.plot(temperature_mode2,frequency_mode2, label='mode2')
plt.legend()
```
