```python
import xarray as xr

ds  = xr.load_dataset('outfile.intensity_raman_0K.h5')
ds1 = xr.load_dataset('outfile.intensity_raman_50K.h5')
ds2 = xr.load_dataset('outfile.intensity_raman_100K.h5')
ds3 = xr.load_dataset('outfile.intensity_raman_200K.h5')
ds4 = xr.load_dataset('outfile.intensity_raman_300K.h5')
a = ds.intensity.to_series()
b = ds1.intensity.to_series()
c = ds2.intensity.to_series()
d = ds3.intensity.to_series()
e = ds4.intensity.to_series()
```

```python
collect = [a,b,c,d,e]
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()

for ii in(collect):
    ax.plot(ii)
ax.legend(['0K', '50K', '100K', '200K', '300K'])
ax.set_xlabel("Frequency (THz)")
ax.set_ylabel("Intensity")
ax.set_xlim([0,14])
```
